```python
import matplotlib.pyplot as plt
import googlemaps
import itertools
import pandas as pd
import descartes
import geopandas as gpd
from shapely.geometry import Polygon, Point
from scipy.spatial.distance import cdist
import numpy as np
```

```python
def load_addresses(addresses_csv):
    '''
    Function for loading in addresses.
    '''

    addresses = pd.read_csv(addresses_csv)
    addresses['full_address'] = addresses['address'] + ', ' + addresses['city']
    addresses = addresses.drop(columns=['address', 'city'])

    return addresses

def create_permuatations(addresses):
    '''
    Function for creating all permutations of addresses og length 2.
    '''

    permutations_addresses = itertools.permutations(list(addresses['full_address']), 2)

    return permutations_addresses

def get_distances(permutations_addresses):
    '''
    Function for getting distance between origin and destination.
    '''

    distances = []
    for origin, destination in permutations_addresses:
        result = gmaps.distance_matrix(origin, destination, mode='driving')["rows"][0]["elements"][0]["distance"]["value"]

        distances.append({'origin': origin, 'destination': destination, 'distance': result})

    return distances

def get_lat_lon(addresses):
    '''
    Function for getting latitude and longitude for addresses.
    '''

    geocodes = {'address': [], 'lat': [], 'lon': []}
    for address in list(addresses['full_address']):
        geocode = gmaps.geocode(address)
        geocodes['address'].append(address)
        geocodes['lat'].append(geocode[0]['geometry']['location']['lat'])
        geocodes['lon'].append(geocode[0]['geometry']['location']['lng'])

    return geocodes

```
